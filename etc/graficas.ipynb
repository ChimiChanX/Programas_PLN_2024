```python
#CATEGORIZAMOS DATOS DE LA COLUMNA EDAD_ANOS EN CUATRO CATEGOR칈AS
#Ni침os de < 12 a침os
#Adolescentes   >=12 y < 20
#Adultos >= 20 y <60
#Adultos mayores >= 60
ni침os = df[df["EDAD_ANOS"]<12].count()
#ni침os = len(ni침os) No me funcion칩 游뱂
ni침os = ni침os["EDAD_ANOS"]
print("N칰mero de ni침os: {}".format(ni침os))

adolescentes = df[(df["EDAD_ANOS"]>=12) & (df["EDAD_ANOS"] < 20)].count()
adolescentes = adolescentes["EDAD_ANOS"]
print("N칰mero de adolescentes: {}".format(adolescentes))

adultos = df[(df["EDAD_ANOS"]>=20) & (df["EDAD_ANOS"] < 60)].count()
adultos = adultos["EDAD_ANOS"]
print("N칰mero de Adultos: {}".format(adultos))

adultosm = df[(df["EDAD_ANOS"]>=60)].count()
adultosm = adultosm["EDAD_ANOS"]
print("N칰mero de Adultos Mayores: {}".format(adultosm))
print("-- VERIFICANDO TOTALES --")
print("* Totales por  Categor칤as: ", ni침os+adolescentes+adultos+adultosm)
print("* Total de Pacientes del Dataset:", df["EDAD_ANOS"].count())
```

```python
# Graficamos por categor칤as de la columna EDAD_ANOS
# Creamos dos objetos fig y ax
fig, ax = plt.subplots()
# Definimos colores para cada barra
colores = ['blue', 'orange', 'green', 'red']
# Creamos barras individuales
ax.bar('Ni침os', ni침os, 0.5, label='Ni침os', color=colores[0])
ax.bar('Adolescentes', adolescentes, 0.5, label='Adolescentes', color=colores[1])
ax.bar('Adultos', adultos, 0.5, label='Adultos', color=colores[2])
ax.bar('Adultos mayores', adultosm, 0.5, label='Adultos mayores', color=colores[3])
ax.set_ylabel('Cantidad')
ax.set_title('Distribuci칩n de Pacientes de Dengue por Edad')
# Agregar las cantidades encima de cada barra
for i, v in enumerate([ni침os, adolescentes, adultos, adultosm]):
    ax.annotate(str(v), xy=(i, v), xytext=(0, 3), textcoords="offset points", ha='center')
# Colocar la leyenda en la esquina superior izquierda
ax.legend(loc='upper left')
# Mostrar gr치fica de barras
plt.show()

```

```python
# Agregamos una gr치fica de barras del campo EDAD_ANOS del dataset
# Categorizar los casos por grupos de edad
# Ni침os menores de 12 a침os
# Adolescentes de 12 y menores de 20 a침os
# Adultos =>20 y <60
# Adulto mayor >= 60
ni침os = df[df["EDAD_ANOS"] < 12]["HEMORRAGICOS"].value_counts()
adolescentes = df[(df["EDAD_ANOS"] >= 12) & (df["EDAD_ANOS"] < 20)]["HEMORRAGICOS"].value_counts()
adultos = df[(df["EDAD_ANOS"] >= 20) & (df["EDAD_ANOS"] < 60)]["HEMORRAGICOS"].value_counts()
adultos_mayores = df[df["EDAD_ANOS"] >= 60]["HEMORRAGICOS"].value_counts()

# Crear la figura
plt.figure(figsize=(10, 6))

# Graficar los datos
ni침os = plt.bar([0, 1], ni침os, color='blue', alpha=0.7, width=0.2, label='Ni침os')
adolescentes_bars = plt.bar([0.2, 1.2], adolescentes, color='orange', alpha=0.7, width=0.2, label='Adolescentes')
adultos_bars = plt.bar([0.4, 1.4], adultos, color='green', alpha=0.7, width=0.2, label='Adultos')
adultos_mayores_bars = plt.bar([0.6, 1.6], adultos_mayores, color='red', alpha=0.7, width=0.2, label='Adultos Mayores')

# Configurar la gr치fica
plt.title('Distribuci칩n de Casos de Dengue Hemorr치gico y Cl치sico por Edad')
plt.xlabel('Tipo de Dengue')
plt.ylabel('N칰mero de casos')
plt.xticks([0.3, 1.3], ['Dengue Cl치sico', 'Dengue Hemorr치gico'])
plt.legend()
plt.tight_layout()

# Agregar etiquetas de n칰mero a cada barra
for bars in [ni침os, adolescentes_bars, adultos_bars, adultos_mayores_bars]:
    for bar in bars:
        height = bar.get_height()
        plt.annotate('{}'.format(height),
                     xy=(bar.get_x() + bar.get_width() / 2, height),
                     xytext=(0, 3),
                     textcoords="offset points",
                     ha='center', va='bottom')

# Mostrar la gr치fica
plt.show()
```

```python
# Contar la cantidad de casos de dengue asignados a cada entidad
casos_por_entidad = df['ENTIDAD_ASIG'].value_counts().sort_index()

# Crear la gr치fica de barras
plt.figure(figsize=(12, 6))
casos_por_entidad.plot(kind='bar', color='green')

# Configurar la gr치fica
plt.title('Comportamiento del Dengue  por Entidad'+"\n" +"Febrero 19 de 2024")
plt.xlabel('Entidad de la Rep칰blica Mexicana')
plt.ylabel('N칰mero de casos')
plt.xticks(rotation=45)  # Rotar las etiquetas del eje x para mayor legibilidad
plt.grid(axis='y')  # Mostrar l칤neas de la cuadr칤cula solo en el eje y

# Agregar etiquetas de n칰meros a cada barra
for index, value in enumerate(casos_por_entidad):
    plt.text(index, value, str(value), ha='center', va='bottom')

# Mostrar la gr치fica
plt.tight_layout()
plt.show()




```

```python
# N칰mero de hombres y mujeres en el dataset
# Contar los valores de la columna "Sexo"
numporSexo = df["SEXO"].value_counts()
print(numporSexo)
# Crear la gr치fica de pastel
plt.figure(figsize=(4, 4))
plt.pie(numporSexo, labels=['Hombres', 'Mujeres'], autopct='%1.1f%%', colors=['skyblue', 'lightcoral'], startangle=90)
plt.title('Distribuci칩n Pacientes de Dengue Clasificado por Hombres y Mujeres')
# Para que el gr치fico de pastel sea circular
plt.axis('equal')  
plt.show()
```

```python
# Contar el n칰mero de casos por categor칤a de HEMORRAGICOS
# Suponiendo que 1 representa la presencia de la condici칩n
hemorragicos = df[df["HEMORRAGICOS"] == 1]
num_hemorragicos = len(hemorragicos)
print("N칰mero de casos con hemorragias:", num_hemorragicos)

# Repite el mismo proceso para las otras variables
diabetes = df[df["DIABETES"] == 1]
num_diabetes = len(diabetes)
print("N칰mero de casos con diabetes:", num_diabetes)

hipertension = df[df["HIPERTENSION"] == 1]
num_hipertension = len(hipertension)
print("N칰mero de casos con hipertensi칩n:", num_hipertension)

ulc_peptica = df[df["ENFERMEDAD_ULC_PEPTICA"] == 1]
num_ulc_peptica = len(ulc_peptica)
print("N칰mero de casos con enfermedad ulceropeptica:", num_ulc_peptica)

enfermedad_renal = df[df["ENFERMEDAD_RENAL"] == 1]
num_enfermedad_renal = len(enfermedad_renal)
print("N칰mero de casos con enfermedad renal:", num_enfermedad_renal)

inmunosupr = df[df["INMUNOSUPR"] == 1]
num_inmunosupr = len(inmunosupr)
print("N칰mero de casos con inmunosupresi칩n:", num_inmunosupr)

cirrosis_hepatica = df[df["CIRROSIS_HEPATICA"] == 1]
num_cirrosis_hepatica = len(cirrosis_hepatica)
print("N칰mero de casos con cirrosis hep치tica:", num_cirrosis_hepatica)

embarazo = df[df["EMBARAZO"] == 1]
num_embarazo = len(embarazo)
print("N칰mero de casos de pacientes embarazadas:", num_embarazo)
```

```python
fig, ax = plt.subplots()
ax.bar('Hemorragias', num_hemorragicos, 0.5, label='Hemorragias')
ax.bar('Diabetes', num_diabetes, 0.5, label='Diabetes')
ax.bar('Hipertensi칩n', num_hipertension, 0.5, label='Hipertensi칩n')
ax.bar('Ulceropeptica', num_ulc_peptica, 0.5, label='Enfermedad Ulceropeptica')
ax.bar('Enfermedad Renal', num_enfermedad_renal, 0.5, label='Enfermedad Renal')
ax.bar('Inmunosupresi칩n', num_inmunosupr, 0.5, label='Inmunosupresi칩n')
ax.bar('Cirrosis Hep치tica', num_cirrosis_hepatica, 0.5, label='Cirrosis Hep치tica')
ax.bar('Embarazo', num_embarazo, 0.5, label='Embarazo')

ax.set_xlabel('N칰mero de Casos')
ax.set_ylabel('Condici칩n M칠dica')
ax.set_title('N칰mero de Casos de Dengue por Condici칩n M칠dica')
# Agregar las cantidades encima de cada barra
for i, v in enumerate([num_hemorragicos, num_diabetes, num_hipertension, num_ulc_peptica, num_enfermedad_renal,num_inmunosupr, num_cirrosis_hepatica, num_embarazo]):
    ax.annotate(str(v), xy=(i, v), xytext=(0, 3), textcoords="offset points", ha='center')
# a 45 grados
plt.xticks(rotation=45, ha='right')
# Colocar la leyenda en la esquina superior derecha
ax.legend(loc='upper right')
plt.show()
```
