```python
import pandas as pd

import numpy as np
 
```

```python
# Asegúrate de colocar la ruta correcta del archivo CSV
file_path = 'dengue_abierto.csv'
dengue_df = pd.read_csv(file_path)
```

```python
# Punto 2: Revisar las primeras filas del DataFrame
dengue_df.head()
```

```python
# Punto 3: Verificar y manejar los datos faltantes
dengue_df.isnull().sum()
# Si es necesario, manejar los valores NaN, por ejemplo, usando fillna() o dropna()
```

```python
# Punto 4: Obtener información estadística básica sobre variables numéricas
dengue_df.describe()
```

```python
# Punto 5: Identificar variables categóricas
categorical_variables = dengue_df.select_dtypes(include=['object']).columns
categorical_variables
```

```python
# Punto 6: Obtener las tres medidas de tendencia central de las variables numéricas
mean_values = dengue_df.mean()
mode_values = dengue_df.mode().iloc[0]
std_deviation = dengue_df.std()
```

```python
# Punto 7: Contar el número de casos por categoría para cada variable categórica
cases_by_category = dengue_df.groupby(categorical_variables).size()
```

```python
# Punto 8: Obtener el número de casos de dengue clásico y dengue hemorrágico
cases_by_dengue_type = dengue_df['tipo_dengue'].value_counts()
```

```python
# Punto 9: Obtener el número de pacientes que fallecieron por cada tipo de dengue y por categoría de edad
deaths_by_dengue_type = dengue_df.groupby(['tipo_dengue', 'categoria_edad'])['fallecido'].sum()
```

```python
# Punto 10: Visualizar la distribución de casos de dengue por categoría usando gráficos de barras
import matplotlib.pyplot as plt
import seaborn as sns

sns.countplot(x='categoria_edad', data=dengue_df, hue='tipo_dengue')
plt.title('Distribución de casos de dengue por categoría')
plt.show()
```

```python
# Punto 11: Visualizar el número de casos de dengue a lo largo del tiempo
# (asumiendo que hay una columna 'fecha' en el DataFrame)
dengue_df['fecha'] = pd.to_datetime(dengue_df['fecha'])
cases_over_time = dengue_df.groupby(dengue_df['fecha'].dt.month)['casos'].sum()
cases_over_time.plot(kind='bar')
plt.title('Número de casos de dengue a lo largo del tiempo')
plt.xlabel('Mes')
plt.ylabel('Número de casos')
plt.show()
```

```python
# Punto 12: Visualizar la distribución de casos de dengue por estados de la república mexicana
sns.countplot(x='estado', data=dengue_df)
plt.title('Distribución de casos de dengue por estados')
plt.xticks(rotation=45, ha='right')
plt.show()
```
